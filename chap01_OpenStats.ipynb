```python
import pandas as pd
data_set = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/csv/migraine.csv')
data_set.head()
```

```python
treatment = (data_set['group'] == 'treatment').sum()
treatment_pain_free = ((data_set['group'] == 'treatment') & (data_set['pain_free'] == 'yes')).sum()


print(f"(a) Percentage of pain free patients is: {round((treatment_pain_free / treatment) * 100,2)}%")

```
