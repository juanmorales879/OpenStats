
# Exercise 1.1

(a) What percent of patients in the treatment group were pain free 24 hours after receiving acupuncture?
(b) What percent were pain free in the control group?
(c) In which group did a higher percent of patients become pain free 24 hours after receiving acupuncture?
(d) Your findings so far might suggest that acupuncture is an effective treatment for migraines for all people
who suffer from migraines. However, this is not the only possible conclusion that can be drawn based
on your findings so far. What is one other possible explanation for the observed difference between the
percentages of patients that are pain free 24 hours after receiving acupuncture in the two groups?

```python
import pandas as pd
data_acupunture = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/csv/migraine.csv')
data_sinus = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/csv/sinusitis.csv')
```

```python
treatment = (data_acupunture['group'] == 'treatment').sum()
treatment_pain_free = ((data_acupunture['group'] == 'treatment') & (data_acupunture['pain_free'] == 'yes')).sum()
control = (data_acupunture['group'] == 'control').sum()
control_pain_free = ((data_acupunture['group'] == 'control') & (data_acupunture['pain_free'] == 'yes')).sum()


print(f"(a) Percentage of pain free patients in treatment group is: {round((treatment_pain_free / treatment) * 100,2)}%")
print(f"(b) Percentage of pain free patients in control group is: {round((control_pain_free / control) * 100,2)}%")
print("(c) Treatment group had a higher percentage of pain free patients")
print("(d) It could also be due to chance only")

```

We can use data matrix for organizing data, the name for a row is a case or **observational** unit. The columns represents characteristics, called **variables**.

Data that takes a wide range of numerical values and is sensible to add, substract or take average is called **numerical**.

Numerical data can be **continuous**, such as unemployment data which can take any range of numbers or **discrete** sush as population count, which can only take whole non negative values.

Variables can also be **categorical** such as the states and can be divided into **ordinal** like degrees where there is a natural ordering or with no special ordering which is called **nominal**.

Scatterplots ar eone type of graph used to study the relationship between two variables. When two variables show connections, they are called dependent.

Downward trends are said to be negatively associated and upwards are reffered to as positively associated.

If two variables are not associated, then they are said to be independent.



Researchers may perform two types of data collection:

Observational study: when they collect data in a way that does not directly interfere with how the data arise.

Experiments: when they want to investigate the possibility of a causal connection, when individuals a randomly assigned to a group the experiment is called a randomized experiment.
